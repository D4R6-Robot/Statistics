```r
# http://www.r-tutor.com/elementary-statistics/numerical-measures/variance
# https://rfriend.tistory.com/110
library(ggplot2)

# 단측 검정과 양측 검정이 다르다. 
```

```r
X <- c(12.61, 11.98, 12.18, 11.52, 12.16, 10.91, 11.22)
n = length(X)
Xn = round(mean(X), 4)
SVX = round(sum((X - mean(X))^2) / (n - 1), 4)

Y <- c(10.68, 11.84, 11.04, 10.31, 11.26, 11.18, 10.96, 12.12)
m = length(Y)
Yn = round(mean(Y), 4)
SVY = round(sum((Y - mean(Y))^2) / (m - 1), 4)

t = round((Xn - Yn) / sqrt((SVX/n) + (SVY/m)), 4)
cat ("검정 통계랑\t: ", t)

# 양측 검정
# t = 2.0278
# H1 : μ≠ μ0, 2 * P( T ≥ΙtΙ)
p <- round(2 * (1 - pt(q = t, df = 6)), 4)
cat ("\n유의 확률\t: ", p)

ggplot(data.frame(x = c(-4, 4)), aes(x = x)) + 
stat_function(fun = dt, args = list(df = 6), size = 1) + 
geom_vline(xintercept = -t, linetype = 'dotted', color='blue', size = 1) +
geom_vline(xintercept = t, linetype = 'dotted', color='blue', size = 1) +
geom_area(stat = "function", fun = dt, args = list(df = 6), fill = "red", alpha = 0.5, xlim = c(-4, -t)) +
geom_area(stat = "function", fun = dt, args = list(df = 6), fill = "red", alpha = 0.5, xlim = c(t, 4))
```

```r
Y <- c(0.8, 0.6, 0.3, -0.1, 1.1, -0.2, 0.3, 0.5, 0.5, 0.3)
n = length(Y)
Yn = round(mean(Y), 4)
SVY = round(sum((Y - mean(Y))^2) / (n - 1), 4)

t = round(Yn / (round(sqrt(SVY), 4) / sqrt(n)), 4)
cat ("검정 통계랑\t: ", t)

# 양측 검정
# t = 3.3485
# H1 : μ≠ μ0, 2 * P( T ≥ΙtΙ)
p <- round(2 * (1 - pt(q = t, df = 9)), 4)
cat ("\n유의 확률\t: ", p)

ggplot(data.frame(x = c(-4, 4)), aes(x = x)) + 
stat_function(fun = dt, args = list(df = 9), size = 1) + 
geom_vline(xintercept = -t, linetype = 'dotted', color='blue', size = 1) +
geom_vline(xintercept = t, linetype = 'dotted', color='blue', size = 1) +
geom_area(stat = "function", fun = dt, args = list(df = 9), fill = "red", alpha = 0.5, xlim = c(-4, -t)) +
geom_area(stat = "function", fun = dt, args = list(df = 9), fill = "red", alpha = 0.5, xlim = c(t, 4))
```
