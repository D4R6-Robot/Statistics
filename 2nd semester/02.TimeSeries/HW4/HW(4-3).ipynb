```r
library(fpp2)
library(gridExtra)
```

```r
# 3 [20 점]. 다음의 시계열자료는 {ε_t} 가 N(0,0.5^2)를 따를 때 시계열모형
data <- c(59, 59, 85, 81, 117, 118, 151, 112, 151, 127, 240, 175, 175, 184, 268, 255, 242, 194, 371, 311, 300,
          256, 406, 348, 354, 353, 388, 365, 427, 389, 481, 361, 498, 415, 548, 410, 487, 517, 522, 484, 578,
          594, 557, 576, 582, 706, 639, 601, 667, 750, 671, 650, 710, 886, 663, 737, 742, 945, 716, 802)

ts_data <- ts(data)
print(ts_data)
```

```r
autoplot(ts_data) + geom_point()
```

```r
# (b)1차 차분된 시계열 (1 - B)Z_t의 시계열그림을 그려라.
# 계절성분이 존재하는가? 존재한다면 계절성분을 제거하기 위하여 계절차분을 하라.
```

```r
diff_ts <- diff(ts_data)
p1 <- autoplot(diff_ts) + ylab("DIFF")

diff_ts4 <- diff(diff_ts, lag = 4)
p2 <- autoplot(diff_ts4) + ylab("DIFF, lag = 4")

diff_ts8 <- diff(diff_ts, lag = 8)
p3 <- autoplot(diff_ts8) + ylab("DIFF, lag = 8")

diff_ts12 <- diff(diff_ts, lag = 12)
p4 <- autoplot(diff_ts12) + ylab("DIFF, lag = 12")

grid.arrange(p1, p2, p3, p4,
    layout_matrix = rbind(
        c(1),
        c(2),
        c(3),
        c(4)))
```

```r
# (c) 계절주기가 4 인 계절차분된 시계열 (1 - B^4)Zt의 시계열그림을 그려라. 비정상성이 존재하는지 판단하라.
```

```r
diff_ts <- diff(ts_data)
p1 <- autoplot(diff_ts) + ylab("DIFF")

diff_ts <- diff(ts_data, lag = 4)
p2 <- autoplot(diff_ts) + ylab("seasonal lag = 4")

diff_ts4 <- diff(diff_ts, lag = 4)
p3 <- autoplot(diff_ts4) + ylab("DIFF, lag = 4")

grid.arrange(p1, p2, p3,
    layout_matrix = rbind(
        c(1),
        c(2),
        c(3)))
```
