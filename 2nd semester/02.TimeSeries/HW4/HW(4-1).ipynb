```r
## ARIMA(1, 2, 1) process
set.seed(1)
n <- 200
x <- w <- rnorm(n)
for (t in 13:n) {
    
    x[t] <- 1 * x[t - 12] + w[t]
    
}

plot(x, type = "l")
```

```r
# a. 다음의 R 코드를 이용하여, σ^2 = 2 인 ARIMA(12, 2, 1) 모델로 데이터를 생성해봅시다.
# 생성 과정은  y_1 = 0  에서 시작합니다. Autoregressive 

set.seed(1)

y <- ts(numeric(100))
e <- rnorm(100, mean = 0, sd = 2)

for(i in 2:100){
    y[i] <- 0.6 * y[i - 1] + e[i]
}

print(y)
autoplot(y)
```
