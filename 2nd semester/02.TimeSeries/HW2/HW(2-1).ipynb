```r
# 1. 엑셀로 첨부한 tute.csv 파일을 열어서 내용을 살펴보면, 열이 4개 있는 것을 확인할 수 있다. 열 B부터
# D에는 각각 분기별 판매량 시계열, 광고비, GDP를 의미한다. 판매량은 1981년부터 2005년까지 어떤 작은
# 회사의 분기별 판매량이고, 광고비는 광고하는데 든 예산이며, GDP (gross domestic product)는
# 국내총생산이다. 여기에서 사용하는 시계열은 물가상승률에 맞추어 조정된 것이다. 다음과 같이 데이터를 R로 읽는다. 

tute <- read.csv("tute.csv", header=TRUE)
head(tute)
str(tute)
```

```r
# (a) 열A는 데이터가 관측된 일자를 의미한다. 함수 ts()를 이용하여 열B-D의 데이터를 시계열 자료로 변환하라. 
mytimeseries <- ts(tute[,-1], start=1981, frequency=4)
head(mytimeseries)
str(mytimeseries)
```

```r
# (b) 문제 (a)에서 변환된 데이터의 이름을mytimeseries라 하자. 아래와 같이 ggfortify::autoplot()을
# 이용하여 3개의 각 시계열을 시간 그래프(time plot)로 나타내본다.
library(ggfortify)
autoplot(mytimeseries, facets=FALSE)
```

```r
autoplot(mytimeseries, facets=TRUE, colour='plot_group')
```
