```r
# 6.8 컴퓨터 산업분야에서 놀랄 만한 기술적 진보 중 하나는 하드디스크에 정보를 저장할 수 있는 능력이다.
# 저장에 드는 비용이 지속적으로 감소되었다. 표 6.20 에는 1988 ~ 1998 년 동안 메가바이트당 평균 가격이 주어져 있다.

library(ggplot2) # ggplot
library(ggfortify) # autoplot
library(gridExtra)

#setwd('H:/01.Korea/01.2019-1학기/05.회귀분석 방법론/Report/Chapter6')
setwd('D:/Working/03.Korea/회귀분석/report/Chapter6')
p190_data <- read.table("P190.txt", header = TRUE)
str(p190_data)
nrow(p190_data)
```

```r
# (a) 데이터를 묘사하는 선형 추세가 있는가? 1988 -> 1, 1989 -> 2 와 같은 식으로 입력시키는 새로운 변수 t 를 정의

for(i in 1 : nrow(p190_data)){
    p190_data$t[i] <- i
}

p190_Year_Price_reg <- lm(Price ~ Year, data = p190_data)
p190_Price_t_reg <- lm(Price ~ t, data = p190_data)

p1 <- ggplot(p190_data, aes(Year, Price)) + geom_point() + geom_smooth(method='lm')
p2 <- ggplot(p190_data, aes(t, Price)) + geom_point() + geom_smooth(method='lm')

grid.arrange(p1, p2, ncol=2, nrow=2)
```

```r
p190_Year_Price_reg <- lm(log(Price) ~ log(Year), data = p190_data)
p190_Price_t_reg <- lm(log(Price) ~ log(t), data = p190_data)

p1 <- ggplot(p190_data, aes(log(Year), log(Price))) + geom_point() + geom_smooth(method='lm')
p2 <- ggplot(p190_data, aes(log(t), log(Price))) + geom_point() + geom_smooth(method='lm')

grid.arrange(p1, p2, ncol=2, nrow=2)
```
