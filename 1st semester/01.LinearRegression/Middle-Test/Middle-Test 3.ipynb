```r
# 3. (20점) R에서 제공하는 airmiles 데이터셋은 1937년~1960년까지 비행기 탑승객 의 여행거리가 저장되어 있다. ggplot 으로 선그래프를 작성하시오. 
# (x축:년도, y축:여행거리. airmiles 는 벡터가 아니기 때문에 다음과 같이 벡터로 바꾼 다음 실행한다.  am <- as.numeric(airmiles) )

# https://www.sciencedirect.com/topics/mathematics/time-series-data

# ggplot2 : Create Elegant Data Visualisations Using the Grammar of Graphics
library(ggplot2)
# dplyr: A Grammar of Data Manipulation
library(dplyr)
# gridExtra : Miscellaneous Functions for "Grid" Graphics
library(gridExtra)

# 1. Use vector 벡터 사용하여 dataframe 만들기.
# https://www.programcreek.com/2014/01/initialize-an-empty-vector-in-r/
x = vector() # empty verctor declare
for(i in start(airmiles)[1] : end(airmiles)[1])
    x <- append(x , i)
am <- as.numeric(airmiles) 
df_airmiles <- data.frame(Year = x, Distance = am)
p1 <- df_airmiles %>% ggplot(aes(x = Year, y = Distance)) + geom_line()

# 2. Use sequence 시퀀스 사용하기
# https://www.math.ucla.edu/~anderson/rw1001/library/base/html/seq.html
x = seq(start(airmiles)[1], end(airmiles)[1])
df_airmiles <- data.frame(Year = x, Distance = am)
p2 <- ggplot(df_airmiles, aes(x = Year, y = Distance)) + 
geom_line(color = 'yellow', size = 3) + geom_line(color = 'black', size = 1)

library(ggfortify)
# 3. Use autoplot 3. autoplot 사용하기
# https://cran.r-project.org/web/packages/ggfortify/vignettes/plot_ts.html >> autoplot
p3 <- autoplot(airmiles, ts.colour = 'red', ts.linetype = 'dashed', size=1, alpha=0.2)

# 4. Use fortify
# https://rdrr.io/cran/ggfortify/man/fortify.ts.html >> fortify
p4 <- ggplot(fortify(airmiles), aes(x=Index, y=Data)) +
geom_line(color = 'blue', linetype='dotdash') # 산점도 형태로 뿌리기

grid.arrange(p1, p2, p3, p4, ncol=2, nrow=2)
```
