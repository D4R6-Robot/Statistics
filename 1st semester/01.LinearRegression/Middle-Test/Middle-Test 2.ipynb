```r
# 2.(20점) R에서 제공하는 mtcars 데이터셋의 gear 데이터를 가지고 기어수별 빈도에 대해 ggplot 으로 막대 그래프를 작성하시오.

# ggplot2 : Create Elegant Data Visualisations Using the Grammar of Graphics
library(ggplot2)
# dplyr: A Grammar of Data Manipulation
library(dplyr)
# gridExtra : Miscellaneous Functions for "Grid" Graphics
library(gridExtra)

mt_car <- data.frame(mtcars) # mtcars dataframe 만들기.

p1 <- ggplot(mt_car, aes(gear, colour=gear)) +               # gear 기준
geom_histogram(color="black", fill="white", binwidth=0.5) +  # histogram 만들기
geom_vline(aes(xintercept=mean(gear)),                       # gear 평균 기준선 vline 긋기
            color="blue", linetype="dashed", size=1)         # vline type 설정.

p2 <- ggplot(mt_car, aes(gear, colour=gear)) +               # gear 기준
geom_histogram(aes(y=..density..), color="black", fill="yellow", binwidth = 0.5) + # histogram 만들기 (st normal dist density 형태로 줄이기)
geom_density(alpha=.2, fill="blue")                          # dinsity 색 칠하기

grid.arrange(p1, p2, ncol=2, nrow=2)
```
