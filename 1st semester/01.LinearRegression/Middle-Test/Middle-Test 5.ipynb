```r
# 5.(20점)iris 데이터셋에서 Species의 정보에 따른 Sepal.Length, Sepal.Width(꽃받침의 길이, 폭)의 분포를 살펴보고 해석하시오.

# ggplot2 : Create Elegant Data Visualisations Using the Grammar of Graphics
library(ggplot2)
# dplyr: A Grammar of Data Manipulation
library(dplyr)
# gridExtra : Miscellaneous Functions for "Grid" Graphics
library(gridExtra)

# https://medium.com/@gscheithauer/how-to-add-number-of-observations-to-a-ggplot2-boxplot-b22710f7ef80
stat_length_data <- function(y, upper_limit = max(iris$Sepal.Length) * 1.15) {
  return( 
    data.frame(
      y = upper_limit,
      label = paste('갯수 =', length(y), '\n',
                    '평균 =', mean(y), '\n',
                    '분산 =', round(var(y), 4), '\n',
                    '중앙 값 =', median(y), '\n',
                    '사분 범위 =', IQR(y), '\n')
    )
  )
}

ggplot(data=iris, aes(x = Species, y = Sepal.Length, fill=Species)) + 
    geom_boxplot(
        # custom outliers
        outlier.colour="red",
        outlier.fill="red",
        outlier.size=2
    ) +
    geom_hline(yintercept = mean(iris$Sepal.Length), color="lightblue", linetype="dashed", size=1) +
    stat_summary(
        fun.data = stat_length_data, 
        geom = "text", 
        hjust = 0.5,
        vjust = 0.9
    )

stat_width_data <- function(y, upper_limit = max(iris$Sepal.Width) * 1.15) {
  return( 
    data.frame(
      y = upper_limit,
      label = paste('\n\n\n',
                    '갯수 =', length(y), '\n',
                    '평균 =', mean(y), '\n',
                    '분산 =', round(var(y), 4), '\n',
                    '중앙 값 =', median(y), '\n',
                    '사분 범위 =', IQR(y), '\n')
    )
  )
}

ggplot(data=iris, aes(x = Species, y = Sepal.Width, fill=Species)) + 
    geom_boxplot(
        # custom outliers
        outlier.colour="red",
        outlier.fill="red",
        outlier.size=2
    ) +
    geom_hline(yintercept = mean(iris$Sepal.Width), color="lightblue", linetype="dashed", size=1) +
    stat_summary(
        fun.data = stat_width_data, 
        geom = "text", 
    )
```
