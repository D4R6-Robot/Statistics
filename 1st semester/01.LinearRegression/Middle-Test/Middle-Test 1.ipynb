```r
# 1.(20점) R에서 제공하는 state.x77 데이터셋의 Income, Illiteracy 데이터를 가지고 ggplot 으로 산점도를 작성하시오.

# ggplot2 : Create Elegant Data Visualisations Using the Grammar of Graphics
library(ggplot2)
# dplyr: A Grammar of Data Manipulation
library(dplyr)
# gridExtra : Miscellaneous Functions for "Grid" Graphics
library(gridExtra)

state <- data.frame(state.x77)
str(state)

p1 <- state %>% ggplot(aes(x=Income, y=Illiteracy)) + # state.x77 데이터를 Income, Illiteracy 기준으로
    geom_point() + # 산점도 형태로 뿌리기
    ggtitle("Income & Illiteracy correlation") + # Title 이름 설정 후
    theme(plot.title = element_text(size=14,face="bold",color="blue", hjust=0.5)) # 원하는 테마 입히기.

p2 <- state %>% ggplot(aes(x=Income, y=Illiteracy)) + # state.x77 데이터를 Income, Illiteracy 기준으로
    geom_point() + # 산점도 형태로 뿌리기
    ggtitle("Income & Illiteracy correlation") + # Title 이름 설정 후
    theme(plot.title = element_text(size=14,face="bold",color="blue", hjust=0.5)) + # 원하는 테마 입히기.
    geom_vline(xintercept = mean(state$Income) ,linetype="dashed", color="red") + # X축 - Income 의 평균 기준 선으로 잡기
    geom_hline(yintercept = mean(state$Illiteracy), linetype="dashed", color="red") # Y축 - Illiteracy 의 평균 기준 선으로 잡기

p3 <- state %>% ggplot(aes(x=Illiteracy, y=Income)) + # state.x77 데이터를 Income, Illiteracy 기준으로
    geom_point() + # 산점도 형태로 뿌리기
    ggtitle("Illiteracy & Income correlation") + # Title 이름 설정 후
    theme(plot.title = element_text(size=14,face="bold",color="blue", hjust=0.5)) # 원하는 테마 입히기.

p4 <- state %>% ggplot(aes(x=Illiteracy, y=Income)) + # state.x77 데이터를 Income, Illiteracy 기준으로
    geom_point() + # 산점도 형태로 뿌리기
    ggtitle("Illiteracy & Income correlation") + # Title 이름 설정 후
    theme(plot.title = element_text(size=14,face="bold",color="blue", hjust=0.5)) + # 원하는 테마 입히기.
    geom_vline(xintercept = mean(state$Illiteracy) ,linetype="dashed", color="red") + # X축 - Income 의 평균 기준 선으로 잡기
    geom_hline(yintercept = mean(state$Income), linetype="dashed", color="red") # Y축 - Illiteracy 의 평균 기준 선으로 잡기

grid.arrange(p1, p2, p3, p4, ncol=2, nrow=2)
```
